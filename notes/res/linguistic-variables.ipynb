```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

```python
import matplotlib

matplotlib.rc('xtick', labelsize=15);
matplotlib.rc('ytick', labelsize=15);
```

```python
fig, ax = plt.subplots()

age_u = np.arange(80)

ax.set_xticks([30, 50])
ax.set_yticks([0, 1])
ax.set_ylabel('$\mu_{y/o}(x)$', fontsize=15)
ax.set_xlabel('$x[years]$', fontsize=15)

young = fuzz.trapmf(age_u, [0, 0, 30, 50])
old = fuzz.trapmf(age_u, [30, 50, 200, 200])

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.plot(young, label='young')
plt.plot(old, label='old')

plt.legend()
plt.tight_layout()

plt.savefig('age.png', dpi=224)
```

```python
fig, ax = plt.subplots()

size_u = np.arange(140, 200)

ax.set_xticks([150, 190])
ax.set_yticks([1])
ax.set_ylabel('$\mu_{s/m/t}(x)$', fontsize=15)
ax.set_xlabel('$x[cm]$', fontsize=15)

small = fuzz.trapmf(size_u[:21], [0, 0, 150, 160])
medium = fuzz.trapmf(size_u[10:51], [150, 160, 180, 190])
tall = fuzz.trapmf(size_u[40:], [180, 190, 200, 200])

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.plot(size_u[:21], small, label='small')
plt.plot(size_u[10:51], medium, label='medium')
plt.plot(size_u[40:], tall, label='tall')

plt.legend()
plt.tight_layout()

plt.savefig('size.png', dpi=224)
```

```python
# Trapezoid <a, b, c, d> 
fig, ax = plt.subplots()

u = np.arange(11)
a, b, c, d = 2, 4, 6, 8

ax.set_xticks([a, b, c, d])
ax.set_xticklabels(['a', 'b', 'c', 'd'])
ax.set_yticks([0, 1])
ax.set_ylabel('$\mu(x)$', fontsize=15)
ax.set_xlabel('$x$', fontsize=15)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.plot(u, fuzz.trapmf(u, [a, b, c, d]))

plt.savefig('tropezoid.png', dpi=224)
```

```python
# Triangular <a, b, b, d>
fig, ax = plt.subplots()

u = np.arange(11)
a, b, d = 2, 5, 8

ax.set_xticks([a, b, d])
ax.set_xticklabels(['a', 'b', 'd'])
ax.set_yticks([0, 1])
ax.set_ylabel('$\mu(x)$', fontsize=15)
ax.set_xlabel('$x$', fontsize=15)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.plot(u, fuzz.trapmf(u, [a, b, b, d]))
plt.savefig('triangular.png', dpi=224)
```

```python
# Singleton (a, 1) y (b, 0.5)
fig, ax = plt.subplots()

u = np.arange(11)
a, b = 3, 6

ax.set_xticks([a, b])
ax.set_xticklabels(['a', 'b'])
ax.set_yticks([0, 1])
ax.set_ylabel('$\mu(x)$', fontsize=15)
ax.set_xlabel('$x$', fontsize=15)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

singleton = np.zeros(11)
singleton[a] = 1
singleton[b] = 0.5

plt.stem(singleton, markerfmt=' ', basefmt=' ')
plt.savefig('singleton.png', dpi=224)
```

```python
# Gaussian <mean, sigma>
fig, ax = plt.subplots()

u = np.arange(11)
mu, sigma = 5, 2

ax.set_yticks([0, 1])
ax.set_ylabel('$\mu(x)$', fontsize=15)
ax.set_xlabel('$x$', fontsize=15)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.plot(u, fuzz.gaussmf(u, mu, sigma))
plt.savefig('gaussian.png', dpi=224)
```
