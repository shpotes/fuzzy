```python
from src.utils import *

sim, (holes, piles, wells), mood = get_expert_model(t_norm=t_drastic, s_norm=s_drastic, with_variables=True)

x, y = np.meshgrid(np.arange(-4, 4), np.arange(0, 4))
z = np.zeros_like(x)

# Loop through the system 21*21 times to collect the control surface
for i in range(0, 4):
    for j in range(-4, 4):
        sim.input['holes'] = x[i, j]
        sim.input['wells'] = y[i, j]
        sim.input['piles'] = 0
        
        sim.compute()
        z[i, j] = sim.output['mood']

# Plot the result in pretty 3D with alpha blending
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Required for 3D plotting

fig = plt.figure(dpi=224)
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap='magma',
                       linewidth=0.4, antialiased=True)
ax.set_xlabel('piles')
ax.set_ylabel('holes')
ax.set_zlabel('score')

cset = ax.contourf(x, y, z, zdir='z', offset=-2.5, cmap='magma', alpha=0.5)
cset = ax.contourf(x, y, z, zdir='x', offset=3, cmap='magma', alpha=0.5)
cset = ax.contourf(x, y, z, zdir='y', offset=3, cmap='magma', alpha=0.5)

ax.view_init(30, 200)
```
