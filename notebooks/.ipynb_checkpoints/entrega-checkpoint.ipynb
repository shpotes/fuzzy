```python
def run_simulation(expert_config):
    env = TetrisEnv()
    
    # para hacer una comparacion justa en este notebook
    # se fijó la generacion de las fichas
    env.seed(42)
    
    env.reset()
    
    state = Node(env)
    cumulative_reward = 0
    ends = False
    boards = []
    
    while not ends:
        # dentro de expert_actions tambien se realiza .prop (ver src.engine)
        state, actions = state.expert_actions(expert_config)
        for act in actions:
            board, reward, ends, _ = env.step(act)
            cumulative_reward += reward
            boards.append(board.T)
            
    return boards, cumulative_reward
```

```python
expert_config = {
    't_norm': utils.t_lukasiewicz, 
    's_norm': utils.s_lukasiewicz,
    'deffuz': 'centroid'
}

boards, cumulative_reward = run_simulation(expert_config)
print(f'el juego duró {len(boards)} fichas y se el se removieron {cumulative_reward} filas')
```
