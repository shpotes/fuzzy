```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl 
import matplotlib.pyplot as plt
```

```python
def get_fuzzy_variables():
    

    piles = ctrl.Antecedent(np.arange(-4, 5), 'piles')
    piles['nice'] = fuzz.membership.trapmf(piles.universe, [-4, -4, 0, 1])
    piles['not_that_bad'] = fuzz.membership.trimf(piles.universe, [0, 1, 2]) #
    piles['wtf'] = fuzz.membership.trapmf(piles.universe, [1, 2, 6, 6])

    wells = ctrl.Antecedent(np.arange(-4, 4), 'wells')
    wells['removed'] = fuzz.membership.trapmf(wells.universe, [-4, -4, 0, 1]) #
    wells['accumulate_more'] = fuzz.membership.trimf(wells.universe, [0, 2, 5])

    mood = ctrl.Consequent(np.arange(15), 'mood')
    mood[':)'] = fuzz.membership.trimf(mood.universe, [0, 2, 3])
    mood[':|'] = fuzz.membership.trapmf(mood.universe, [2, 4, 5, 7])
    mood[':('] = fuzz.membership.trimf(mood.universe, [6, 8, 10])
    mood[':(('] = fuzz.membership.trimf(mood.universe, [9, 10, 15])

    return (holes, piles, wells), mood
```

```python
holes = ctrl.Antecedent(np.arange(-2, 6, 0.1), 'holes')

no = fuzz.membership.dsigmf(holes.universe, 0, 2, 0.5, 4)
no = np.where(no >= 0, no, 0)
no = no * 1 / no.max()

mid = fuzz.membership.dsigmf(holes.universe, 1, 2, 2, 4)
mid = np.where(mid >= 0, mid, 0)
mid = mid * 1 / mid.max()

holes['none'] = no #fuzz.membership.trimf(holes.universe, [-2, 0, 0])
holes['medium'] = mid #fuzz.membership.trimf(holes.universe, [0, 1, 3]) #
holes['high'] = fuzz.membership.sigmf(holes.universe, 2, 2)

plt.figure(dpi=220)
holes.view()
plt.show()
```

```python
piles = ctrl.Antecedent(np.arange(-4, 5, 0.1), 'piles')

#piles['nice'] = fuzz.membership.trapmf(piles.universe, [-4, -4, 0, 1])
piles['nice'] = fuzz.membership.sigmf(piles.universe, 0.5, -4)
piles['could be better'] = fuzz.membership.gaussmf(piles.universe, 1, 0.5)
#piles['wtf'] = fuzz.membership.trapmf(piles.universe, [1, 2, 5, 5])
piles['bad'] = fuzz.membership.sigmf(piles.universe, 1.5, 3)

piles.view()
```

```python
wells = ctrl.Antecedent(np.arange(-4, 4, 0.1), 'wells')
#wells['removed2'] = fuzz.membership.trapmf(wells.universe, [-4, -4, 0, 1]) #
wells['remove'] = fuzz.membership.sigmf(wells.universe, 0, -3) #
wells['accumulate more'] = fuzz.membership.sigmf(wells.universe, 1, 3)
#fuzz.membership.trimf(wells.universe, [0, 2, 5])

wells.view()
```

```python
mood = ctrl.Consequent(np.arange(11), 'mood')

mood = ctrl.Consequent(np.arange(15, step=0.1), 'mood')

mood['good'] = fuzz.membership.gaussmf(mood.universe, 2, 1)
mood['half-hearted'] = fuzz.membership.gbellmf(mood.universe, 2, 3, 5)
mood['bad'] = fuzz.membership.gaussmf(mood.universe, 8, 1)
mood['crisis'] = fuzz.membership.sigmf(mood.universe, 12, 2)

mood.view()
```
